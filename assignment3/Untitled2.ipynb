```python
from konlpy.tag import Komoran

def change(text):
    komoran = Komoran()
    results = []

    # 없는 태그
    tag_map = {
        'EF': 'EM',
        'EC': 'EM',
        'JC': 'JKB',
        'VCN': 'VCP'
    }

    lines = text.strip().split('\n')

    for line in lines:
        line = line.strip()
        if not line:
            continue

        morphs = komoran.pos(line)


        formatted = "+".join([
            f"{word}/{tag_map.get(tag, tag)}"
            for word, tag in morphs
        ])
        result_line = f"{line}\t{formatted}"
        results.append(result_line)

    return "\n".join(results)
output = change(text)
print(output)
```
